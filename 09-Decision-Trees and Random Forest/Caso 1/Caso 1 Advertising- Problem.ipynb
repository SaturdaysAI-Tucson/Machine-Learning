<h3>Construir un modelo de Regresión Logistica</h3>

```python
# primero asignamos los valores de x y de y, siendo este último nuestro valor que queremos predecir.
X = dataset.iloc[:,[0,1,2,3,6,9,10,11,12]].values
y = dataset.iloc[:,8].values
```

```python
# particionamos el dataset en dos, uno para entrenamiento y el otro para comprobar nuestro modelo.

# tenemos que importar una nueva libreria para hacer el split. 
# dejaremos 80% para el training y 20% para el test.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Escalar los datos
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
X_train = standardScaler.fit_transform(X_train)
X_test = standardScaler.transform(X_test)
```

```python
# iniciar y ajustar el modelo de regresión logistica en el dataset de entrenamiento.

# primero importamos la libreria de Regresion Logistica de SkLearn
from sklearn.linear_model import LogisticRegression
log_rg = LogisticRegression()
log_rg.fit(X_train, y_train)

# Predicción
y_log_rg = log_rg.predict(X_test)
```

<h3>Evaluar Resultados</h3>

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_log_rg)
sns.heatmap(cm,annot=True,fmt='3.0f',cmap="Greens")
plt.title('Confusion matrix para Regresión Logistica', y=1.05, size=15)
```

```python
# Reporte de Clasificación
from sklearn.metrics import classification_report
cr = classification_report(y_test, y_log_rg)
print(cr)
```
